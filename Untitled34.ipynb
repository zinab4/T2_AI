```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers, models
import os
```

```python
train_dir = "/content/drive/MyDrive/T_AI/train"
val_dir = "/content/drive/MyDrive/T_AI/val"

batch_size = 16
img_size = (224, 224)

```

```python
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size=img_size,
    batch_size=batch_size
)

val_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    val_dir,
    image_size=img_size,
    batch_size=batch_size
)

class_names = train_dataset.class_names
print("Detected classes:", class_names)

```

```python
mobilenet_url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

feature_extractor_layer = hub.KerasLayer(
    mobilenet_url,
    input_shape=(224, 224, 3),
    trainable=False
)


print(type(feature_extractor_layer))  # <class 'tensorflow_hub.keras_layer.KerasLayer'>


```

```python
import tensorflow as tf
import tensorflow_hub as hub
from tensorflow.keras import layers, models

def build_model(input_shape=(224, 224, 3), num_classes=2):
    # Load MobileNetV2 without the top classification layer
    base_model = tf.keras.applications.MobileNetV2(
        input_shape=input_shape,
        alpha=1.0,
        include_top=False,  # Exclude the top layer for custom classification
        weights='imagenet',  # Use pre-trained ImageNet weights
        pooling='avg'  # Apply global average pooling
    )

    # Freeze the base model layers
    base_model.trainable = False

    # Build the model with an additional classification layer
    model = tf.keras.Sequential([
        base_model,  # Add MobileNetV2 as base
        layers.Dense(num_classes, activation='softmax')  # Final classification layer
    ])

    return model

# Build the model
model = build_model(input_shape=(224, 224, 3), num_classes=2)

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Display the model summary
model.summary()

```

```python
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=3
)
```

```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import random

# Define the directory where the images are located
image_directory = '/content/drive/MyDrive/T_AI/train/apple'
# Get a list of all image files in the directory (filter to only include image files)
image_files = [f for f in os.listdir(image_directory) if f.endswith('.jpg')]

# If no images are found, print an error message
if not image_files:
    print("No image files found in the directory.")
else:

    random_image_file = random.choice(image_files)
    image_path = os.path.join(image_directory, random_image_file)

    image = tf.keras.preprocessing.image.load_img(image_path)
    image_array = tf.keras.preprocessing.image.img_to_array(image)

    # Plot the original image
    plt.figure(figsize=(6, 6))
    plt.subplot(1, 2, 1)  # First image in a single row
    plt.imshow(image_array.astype('uint8'))
    plt.title("Before Transformation")
    plt.axis('off')  # Hide axes
    image_resized = image.resize((224, 224))

    # Convert the resized image to a numpy array
    image_resized_array = tf.keras.preprocessing.image.img_to_array(image_resized)

    # Plot the transformed image
    plt.subplot(1, 2, 2)  # Second image in a single row
    plt.imshow(image_resized_array.astype('uint8'))
    plt.title("After Transformation")
    plt.axis('off')  # Hide axes

    # Display both images
    plt.show()

```
