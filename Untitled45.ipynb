```python
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0, ResNet50, MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Path to the dataset
dataset_path = './CUB_200_2011/images'

# Check if the dataset path exists
if not os.path.exists(dataset_path):
    print("Dataset path not found. Please check the path.")
else:
    print(f"Dataset found at {dataset_path}")

    # Image scaling and augmentation for training
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest',
        validation_split=0.2  # 20% for validation split if no separate validation folder
    )

    # Image scaling only for validation
    validation_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2
    )

    # Create training generator
    train_generator = train_datagen.flow_from_directory(
        dataset_path,
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        subset='training'  # Important: use 'training' subset
    )

    # Create validation generator
    validation_generator = validation_datagen.flow_from_directory(
        dataset_path,
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        subset='validation'  # Important: use 'validation' subset
    )

    # Print class labels to check the categories
    print(train_generator.class_indices)

    # Display first 5 images from the first batch
    images, labels = next(train_generator)

    plt.figure(figsize=(15, 10))

    for i in range(5):
        plt.subplot(1, 5, i+1)
        plt.imshow(images[i])
        class_name = list(train_generator.class_indices.keys())[labels[i].argmax()]
        plt.title(class_name, fontsize=10)
        plt.axis('off')

    plt.tight_layout()
    plt.show()


```

```python


# Assume train_generator and validation_generator are already defined

models = {
    'EfficientNetB0': EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'ResNet50': ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'MobileNetV2': MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
}

history_dict = {}

for model_name, base_model in models.items():
    # Add custom layers on top of base model
    x = base_model.output
    x = tf.keras.layers.GlobalAveragePooling2D()(x)  # Global average pooling to reduce feature maps
    x = tf.keras.layers.Dense(1024, activation='relu')(x)  # Fully connected layer with 1024 units
    predictions = tf.keras.layers.Dense(train_generator.num_classes, activation='softmax')(x)  # Output layer

    model = tf.keras.models.Model(inputs=base_model.input, outputs=predictions)

    # Freeze base model layers to train only custom layers initially
    for layer in base_model.layers:
        layer.trainable = False

    # Compile the model with Adam optimizer and categorical crossentropy loss
    model.compile(optimizer=Adam(learning_rate=0.0001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # Define callbacks for saving best weights and early stopping
    checkpoint = ModelCheckpoint(f'{model_name}_best_model.weights.h5',
                                 save_best_only=True,
                                 save_weights_only=True)
    early_stopping = EarlyStopping(monitor='val_loss',
                                   patience=3,
                                   restore_best_weights=True)

    # Initial training: train only the new layers
    print(f"Training {model_name} (initial)...")
    history = model.fit(
        train_generator,
        epochs=4,
        validation_data=validation_generator,
        callbacks=[checkpoint, early_stopping]
    )
    history_dict[f'{model_name}_initial'] = history
    # Load best weights before fine-tuning
    model.load_weights(f'{model_name}_best_model.weights.h5')

    # Unfreeze first 100 layers for fine-tuning
    for layer in base_model.layers[:100]:
        layer.trainable = True

    # Re-compile with lower learning rate for fine-tuning
    model.compile(optimizer=Adam(learning_rate=0.00001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # Fine-tuning: train both base and new layers
    print(f"Fine-tuning {model_name}...")
    history_finetune = model.fit(
        train_generator,
        epochs=4,
        validation_data=validation_generator,
        callbacks=[checkpoint, early_stopping]
    )
    history_dict[f'{model_name}_finetune'] = history_finetune

# Plot accuracy curves for comparison
plt.figure(figsize=(12, 8))
for model_name, history in history_dict.items():
    plt.plot(history.history['accuracy'], label=f'{model_name} Accuracy')
plt.title('Training Accuracy Comparison')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python



models = {
    'EfficientNetB0': EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'ResNet50': ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3)),
    'MobileNetV2': MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
}

history_dict = {}

for model_name, base_model in models.items():
    # Freeze base model
    base_model.trainable = False

    # Add custom classification head
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    predictions = Dense(train_generator.num_classes, activation='softmax')(x)

    model = Model(inputs=base_model.input, outputs=predictions)
    model.compile(optimizer=Adam(learning_rate=0.0001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    print(f"\nTraining {model_name}...")
    history = model.fit(
        train_generator,
        epochs=2,
        validation_data=validation_generator,
        verbose=1
    )
    history_dict[model_name] = history

# Plot accuracy curves
plt.figure(figsize=(10, 6))
for model_name, history in history_dict.items():
    plt.plot(history.history['accuracy'], label=f'{model_name} Accuracy')
plt.title('Model Accuracy Comparison (Light Training)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the loss curves and accuracy curves for each model

# Define a function to plot loss and accuracy curves
def plot_loss_accuracy_curves(history, model_name):
    # Plot loss curve
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} - Loss Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot accuracy curve
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} - Accuracy Curve')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Example for plotting the loss and accuracy curves for each model
for model_name, history in history_dict.items():
    plot_loss_accuracy_curves(history, model_name)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns

if train_generator.num_classes <= 10:
    # Predict on validation data
    val_preds = model.predict(validation_generator)
    val_preds_classes = np.argmax(val_preds, axis=1)
    val_true_classes = validation_generator.classes

    cm = confusion_matrix(val_true_classes, val_preds_classes)

    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```
